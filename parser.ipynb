```python
#/*******************************/
#/* Return data:                */
#/* [                           */
#/*    Address,                 */
#/*    [                        */
#/*       Data, Data, Data, ... */
#/*    ]                        */
#/* ]                           */
#/*******************************/

def read_hex(hex_filename):
    base_address = 0
    hex_file_data = []

    print("LOAD HEX DATA FILE: " + hex_filename)
    print('-'*66)
    file = open(hex_filename, 'r')#, 0)
    for curr_line in file.readlines():
        print("\t{}".format(curr_line), end='')
        
        # PIC16F84A format       :BBaaAATTDDCC
        #                   BB        number of bytes in line
        #                   aaAA      address to store bytes (divide by 2??)
        #                   TT        00 - program data; 01 - EOF; 04 - extended address
        #                   DD        data
        #                   CC        checksum = inv(BB + aa + AA + TT + DD) + 1
            
        start_code = curr_line[0:1]
        BB = int(curr_line[1:3], 16)
        aaAA = int(curr_line[3:7], 16)
        TT = int(curr_line[7:9], 16)
        DD = curr_line[9:-3]
        CC = int(curr_line[-3:-1], 16)
        print("\t\tBB: {}\n\t\taaAA: {}\n\t\tTT:  {}\n\t\tDD:  {}\n\t\tCC:  {}".format(BB, aaAA, TT, DD, CC))
            
        if start_code == ":":
            curr_CC = 0
            cc_string = ''
            for CC_part in range(1, 9+2*BB, 2):
                _part_value = int(curr_line[CC_part:CC_part + 2], 16)
                cc_string += hex(_part_value) + ' '
                curr_CC += _part_value
            
            print("{}\nCC parts: {}".format('-'*26, cc_string)) 
            curr_CC = ((~curr_CC) & 0xff) + 1    
            
            if curr_CC > 255:
                curr_CC = curr_CC - 256
                
            if CC != curr_CC:
                print("CHECKSUM MISMATCH: {} != {}".format(CC, curr_CC))

            # data type == 0
            if TT == 0:
                print("ADDRESS: {} ".format(base_address + aaAA))
                print("DATA [ {} ]".format(BB))
                hex_file_data.append([hex(BB + aaAA), []])
                
                for data_count in range(0, BB * 2, 2):
                    hex_file_data[len(hex_file_data) - 1][DATA].append(DD[data_count:data_count + 2])
                    aaAA = aaAA + 1
                    
                sys.stdout.write("ADDRESS: {}\n".format(BB + aaAA))

            # data type == 01
            elif TT == 1:
                print("EOF")
                
            # data type == 02 -- Extended segment address type
            elif TT == 2:
                base_address = int(DD, 16) * 16
                print("EXTENDED SEGMENT ADDRESS: {}".format(hex(base_address)))
                
            # data type == 03 -- Start segment address type
            elif TT == 3:
                base_address = int(DD, 16) * 16
                print("START SEGMENT ADDRESS: {}".format(hex(base_address)))
                
            # data type == 04 -- Extended linear address type     
            elif TT == 4:
                base_address= int(DD, 16) * int(math.pow(2, 16))
                print("EXTENDED LINEAR ADDRESS: {}".format(hex(base_address)))
                
            # data type == 05 -- Start linear address type
            elif RecordType == 5:
                base_address = int(DD, 16) * int(math.pow(2, 16))
                print("START LINEAR ADDRESS: {}".format(hex(base_address)))
        print('-'*66)

    file.close()
        
    return hex_file_data
```

```python
PIC_BASE = 14


def _check_str_equality(s1, s2):
    if s1 == s2:
        return True
    return False

def _most_suitable_key(d, opc, check_bits_count):
    chk_opc = opc[:check_bits_count]
    
    for k, v in d.items():
        chk_k_bits = k[:check_bits_count]
            
        if chk_opc == chk_k_bits:
            if k[check_bits_count] in ['x', 'k']:
                return d[k], opc[7:]
    
    return None, None


def _find_command_opcode(opc):
    operation_type = opc[:2]
    
    if operation_type == '00':
        # check equality with 4 operations from "LITERAL AND CONTROL OPERATIONS" block: 
        # [CLRWDT, RETFIE, RETURN, SLEEP]
        check_commands_list = ["CLRWDT", "RETFIE", "RETURN", "SLEEP"]
        for command in check_commands_list:
            if _check_str_equality(opc, PIC16F84A_MNEMONIC_OPCODES[command]):
                return command, opc[7:]
            
        second_to_six_bits = opc[2:6]
        for k, v in PIC16F84A_OPCODES_MNEMONIC.items():
            second_to_six_k_bits = k[2:6]
            if second_to_six_bits == second_to_six_k_bits:
                return PIC16F84A_OPCODES_MNEMONIC[k], opc[7:]
            
    elif operation_type == '01':
        # 100% BIT-ORIENTED FILE REGISTER OPERATIONS
        # take first 4 bits from opc
        four_bits_opc = opc[:4]
        for k, v in PIC16F84A_OPCODES_MNEMONIC.items():
            four_k_bits = k[:4]
            if four_bits_opc == four_k_bits:
                return PIC16F84A_OPCODES_MNEMONIC[k], opc[7:]
            
    elif operation_type == '10':
        # check next 1 bit only
        three_bits_opc = opc[:3]
        for k, v in PIC16F84A_OPCODES_MNEMONIC.items():
            three_k_bits = k[:3]
            if three_bits_opc == three_k_bits:
                return PIC16F84A_OPCODES_MNEMONIC[k], opc[7:]
    else:
        # operation type == 11
        check_bits_count = 4
        
        while check_bits_count <= 6:
            res, res_data = _most_suitable_key(PIC16F84A_OPCODES_MNEMONIC, opc, check_bits_count)
            if res is not None:
                return res, res_data
            
            check_bits_count += 1
            
    
    return "ERROR: {}".format(opc), ""   

def _hex_to_bin(h):
    hb = bin(int(h, 16))[2:]
    return '0' * (PIC_BASE - len(hb)) + hb
    
    
def data_parts_to_opcodes(data_parts_list):
    vhdl_mnemonics_operands = []
    
    for i in range(0, len(data_parts_list), 2):
        pair = data_parts_list[i:i+2]
        pair_little_endian = pair[1] + pair[0]
        pair_opcode = _hex_to_bin(pair_little_endian)
        pic_mnemonic, dt = _find_command_opcode(pair_opcode)

        vhdl_mnemonic = PIC16F84A_VHDL_MNEMONICS[pic_mnemonic]
        
        print(pair_opcode, " --> ", pic_mnemonic, " <= ", dt, " ~ ", vhdl_mnemonic, " <= ", hex(int(dt, 2)))
        vhdl_mnemonics_operands.append([vhdl_mnemonic, dt])
    
    return vhdl_mnemonics_operands
        
def hex_data_parse(hexd):
    data_strings_mnemonics_operands = []
    
    for data_s in hexd:
        addr, d = data_s[0], data_s[1:]
        print(addr, d)
        d_mnem_oper = data_parts_to_opcodes(d[0])
        data_strings_mnemonics_operands += d_mnem_oper
    
    return data_strings_mnemonics_operands
```

```python
VHDL_HEADER = """library IEEE;
use IEEE.STD_LOGIC_1164.ALL;
use IEEE.STD_LOGIC_ARITH.ALL;
use IEEE.STD_LOGIC_UNSIGNED.ALL;

library work;
use work.MPFSM.ALL;\n"""

VHDL_BEHAVIORAL = "\narchitecture Behavioral of {} is"
VHDL_END_BEHAVIORAL = "\nend Behavioral;"

        
def pic_mnem_operand_string_to_vhdl(pic_mnem, pic_operand):
    return '  => {} & "{}", '.format(pic_mnem, pic_operand)


def make_vhdl(vhdl_mnemonics_operands_list, prefix="__CHANGE"):
    vhdl_code = ""
    
    for ind, vhdl_mnem_opnd in enumerate(vhdl_mnemonics_operands_list):
        vhdl_m, vhdl_o = vhdl_mnem_opnd[0], vhdl_mnem_opnd[1]
        codestring = str(ind) + pic_mnem_operand_string_to_vhdl(vhdl_m, vhdl_o)
        vhdl_code += '\t' + codestring + "\n"

    code_formatted = """
    constant {} : type{} := (
    {}
        others => None & None
	);""".format(prefix, prefix, vhdl_code)
    
    code_with_header_end = VHDL_HEADER + VHDL_BEHAVIORAL.format(prefix) + code_formatted + VHDL_END_BEHAVIORAL
    return code_with_header_end
    
```
